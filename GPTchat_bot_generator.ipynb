```python
# Get chatgpt to do some work, adjust prompts or number of requests in prompt if necessary.
# it will usually not give you more than 20 sentences per request, even though we ask it to give us a 100. 
# used prompts
#" please give me 100 tweet text that Twitter bot generates in csv file format",
# "please give me 100 tweet text related to politics that Twitter bot generates in csv file format"
# "please give me 100 tweet text related to any industry that Twitter bot generates in csv file format" 
# "please give me 100 tweet text related to random topic that Twitter bot generates in csv file format",
# "please give me 100 tweet text related to any kind of news that Twitter bot generates in csv file format",
# "please give me 100 tweet text related to Top 50 Most Followed Twitter Accounts that Twitter bot generates in csv file format"

import os
import openai
import json

openai.api_key = '' #remove key 
bigstring = ''
m=0
while m<15:
  response = openai.Completion.create(
    model="text-davinci-003",
    prompt="please give me 100 tweet text that Twitter bot generates in csv file format",
    temperature=0.9,
    max_tokens=4050,
    top_p=1,
    frequency_penalty=0.0,
    presence_penalty=0.6,
    stop=[" Human:", " AI:"]
  )
  response
  json_object = json.loads(str(response))
  text = json_object['choices'][0]['text']
  bigstring = bigstring + text
  m=m+1
```

```python
bigstring = bigstring.split('.')
```

```python
#create a corresponding index for the sentences.
indoix = []
for i in range(len(bigstring)):
  indoix.append(i)
```

```python
#put both lists together in a dictionary
d = {'Indoix':indoix,'Text':bigstring}
```

```python
#put into dataframe and drop 'indoix' column. 
import pandas as pd
df = pd.DataFrame(d)
df = df['Text']
df = df.to_frame()
```

```python
df
```

```python
#get rid of all the \n2 \n3
df = df[df['Text'].str.len()>4]
df = df.reset_index()
df=df['Text'].to_frame()
df
```

```python
#get rid of more trash
import re
newline_symbols='/\r?\n|\r/'
numbers='[0-9]'
df['Text'] = df['Text'].replace(newline_symbols,'',regex=True)
df['Text'] = df['Text'].replace(numbers,'',regex=True)
```
